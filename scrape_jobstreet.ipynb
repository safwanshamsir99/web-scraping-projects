```python
import selenium
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import re
import time
import pandas as pd
import numpy as np
```

## REMINDER: CHANGE PAGE NUM, SPECIALIZATION_ID (if necessary), DF JOB_SPECIALIZATION & FILE NAME

```python
driver = webdriver.Chrome(executable_path='C:\\driver\\chromedriver-win64\\chromedriver.exe')

#Initialize an empty list to store data
job_links_list = []
titles = []
companies = []
locations = []
salaries = []
descriptions = []
level = []
qualifications = []
experiences = []
types = []
registrations = []
sizes = []
industries = []
job_specializations = []


# Loop through the first {} pages
for page_number in range(1, 104):
    page_url = f"https://www.jobstreet.com.my/jobs-in-sales?pg={page_number}"
    
    # Navigate to the page
    driver.get(page_url)
    
    # Find all job links on the page
    job_links = driver.find_elements_by_css_selector('a[href^="/en/job/"]')

    # Extract and print the href attributes
    for link in job_links:
        job_link = link.get_attribute('href')
        job_links_list.append(job_link)
    
# Click the "Next" button to navigate to the next page
time.sleep(5)

for link in job_links_list:
    driver.get(link)
    
    wait = WebDriverWait(driver, 10)
    
        # Extract job information using XPaths
    try:
        job_title = driver.find_element(By.XPATH, '//*[@id="contentContainer"]/div/div/div[1]/div[1]/div[1]/div/div/div[1]/div/div/div[2]/div/div/div/div[1]/h1').text  
    except:
        job_title = "None"
    titles.append(job_title)
        
    try:
        company_name = driver.find_element(By.XPATH, "//*[@id='contentContainer']/div/div/div[1]/div[1]/div[1]/div/div/div[1]/div/div/div[2]/div/div/div/div[2]/span").text
    except:
        company_name = "None"
    companies.append(company_name)
    
    try:
        location = driver.find_element(By.XPATH, "//*[@id='contentContainer']/div/div/div[1]/div[1]/div[1]/div/div/div[2]/div/div/div/div[1]/div/span").text
    except:
        location = "None"
    locations.append(location)
    
    try:
        salary_element = driver.find_element(By.XPATH, "//*[contains(text(), 'MYR')]")
        salary = salary_element.text.strip()
    except:
        salary = "None"
    salaries.append(salary)
    
    
    try:
        job_desc_element = driver.find_element(By.XPATH, "//*[@data-automation='jobDescription']")
        job_desc = job_desc_element.text.strip()
    except:
        job_desc = "None"
    descriptions.append(job_desc)
    
        
    try:
        cl_element = driver.find_element(By.XPATH, "//*[@id='contentContainer']")
        text_content = cl_element.text
        cl_match = re.search(r'Career Level\n(.+)', text_content)
    
        if cl_match:
            career_level = cl_match.group(1).strip()
        else:
            career_level = "None"
    except:
        career_level = "None"
    level.append(career_level)
    
        
    try:
        qualification_element = driver.find_element(By.XPATH, "//*[@id='contentContainer']")
        text_content = qualification_element.text
        qualification_match = re.search(r'Qualification\n(.+)', text_content)
    
        if qualification_match:
            qualification = qualification_match.group(1).strip()
        else:
            qualification = "None"
    except:
        qualification = "None"
    qualifications.append(qualification)
    
        
    try:
        year_element = driver.find_element(By.XPATH, "//*[@id='contentContainer']")
        text_content = year_element.text
        year_match = re.search(r'Years of Experience\n(.+)', text_content)
    
        if year_match:
            year = year_match.group(1).strip()
        else:
            year = "None"
    except:
        year = "None"
    experiences.append(year)
    
        
    try:
        job_type_element = driver.find_element(By.XPATH, "//*[@id='contentContainer']")
        text_content = job_type_element.text
        job_type_match = re.search(r'Job Type\n(.+)', text_content)
    
        if job_type_match:
            job_type = job_type_match.group(1).strip()
        else:
            job_type = "None"
    except:
        job_type = "None"
    types.append(job_type)
    
        
    try:
        registration_no_element = driver.find_element(By.XPATH, "//*[@id='contentContainer']")
        text_content = registration_no_element.text
        registration_no_match = re.search(r'Registration No.\n(.+)', text_content)
    
        if registration_no_match:
            registration_no = registration_no_match.group(1).strip()
        else:
            registration_no = "None"
    except:
        registration_no = "None"
    registrations.append(registration_no)
    
        
    try:
        company_size_element = driver.find_element(By.XPATH, "//*[@id='contentContainer']")
        text_content = company_size_element.text
        company_size_match = re.search(r'Company Size\n(.+)', text_content)
    
        if company_size_match:
            company_size = company_size_match.group(1).strip()
        else:
            company_size = "None"
    except:
        company_size = "None"
    sizes.append(company_size)
    
        
    try:
        industry_element = driver.find_element(By.XPATH, "//*[@id='contentContainer']")
        text_content = industry_element.text
        industry_match = re.search(r'Industry\n(.+)', text_content)
    
        if industry_match:
            industry = industry_match.group(1).strip()
        else:
            industry = "None"
    except:
        industry = "None"
    industries.append(industry)
    
time.sleep(5)
# Close the browser when done
driver.quit()
```

```python
df = pd.DataFrame()

df['Title'] = titles
df['Company'] = companies
df['Location']= locations
df['Salary (MYR)'] = salaries
df['Job Description'] = descriptions
df['Career Level'] = level
df['Qualification'] = qualifications
df['Years of Experience'] = experiences
df['Job Type'] = types
df['Registration No'] = registrations
df['Company Size'] = sizes
df['Industry'] = industries
df['Job Specialization'] = 'Sales/Marketing'
#df['Specialization'] = job_specialization
```
