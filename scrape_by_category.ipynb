## Access subcategories

```python
data=str()
subcat_link = "https://shopee.com.my/Brushes-Beauty-Tools-cat.11000168.11000180"
#for subcat in subcat_links[:1]:
driver.get(subcat_link)
try:
    time.sleep(3)
    close_btn = driver.execute_script('return document.querySelector("#main shopee-banner-popup-stateful").shadowRoot.querySelector("div.home-popup__close-area div.shopee-popup__close-btn")')
    close_btn.click()
except:
    pass
time.sleep(5)
url = driver.current_url

for page in range(9): # pagination [1 page = 60 cards, 17 pages > 1000 cards]
    
    # Define an initial value
    temp_height=0
    while True:
        #Looping down the scroll bar
        driver.execute_script("window.scrollBy(0,1000)")
        #sleep and let the scroll bar react
        time.sleep(5)
        #Get the distance of the current scroll bar from the top
        check_height = driver.execute_script("return document.documentElement.scrollTop || window.pageYOffset || document.body.scrollTop;")
        #If the two are equal to the end
        if check_height==temp_height:
            break
        temp_height=check_height

    data += driver.page_source
    time.sleep(10)

    #navigate to the next page
    next = driver.find_element(By.CSS_SELECTOR,' button.shopee-icon-button.shopee-icon-button--right ')
    driver.execute_script("arguments[0].click();", next)
    time.sleep(5)
```

# BeautifulSoup

```python
soup = BeautifulSoup(data, 'html.parser')
product_cards = soup.find_all('div','col-xs-2-4')

records = []
for card in product_cards:
    name_element = card.find('div','APSFjk cB928k skSW9t').text.strip()
    product_name = str(name_element.encode('ascii','ignore'),'utf-8')
    product_price = card.find('div','JtW3j3 COjcOU').text.strip()
    sales_element = card.find('div','QE5lnM _2pKDjP')
    if sales_element is None:
        product_sold = None
    else:
        product_sold = sales_element.text.strip()
    product_place = card.find('div','gmAE2k').text.strip()
    category_link = 'https://shopee.com.my/Health-Beauty-cat.11000168'
    product_category = '-'.join(category_link.split('/')[-1].split('-')[:-1])
    subcat_link = "https://shopee.com.my/Brushes-Beauty-Tools-cat.11000168.11000180"
    product_subcategory = '-'.join(subcat_link.split('/')[-1].split('-')[:-1])

    product_info = (product_name, product_price, product_sold, product_place, product_category, product_subcategory)

    records.append(product_info)
```

```python
url = "https://shopee.com.my/Brushes-Beauty-Tools-cat.11000168.11000180"
# Split the URL by '/' and take the last element
last_part = url.split('/')[-1]
# Split the last part by '-' and take the first element
result = '-'.join(last_part.split('-')[:-1])
print(result)
```

```python
# Save into CSV file
col = ['Product_Name','Product_Price','Product_Sold','Product_Place','Product_Category','Product_Subcategory']
scrape_data = pd.DataFrame(records,columns=col)

df = pd.read_csv('C:/Users/Acer/Downloads/shopee_data.csv')

# Concatenate 'scrape_data' dataframe with 'df'
df = pd.concat([df, scrape_data], ignore_index=True)

# Write the updated dataframe back to 'shopee_data.csv'
df.to_csv('C:/Users/Acer/Downloads/shopee_data.csv', index=False)
```
