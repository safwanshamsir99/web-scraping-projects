# BeautifulSoup

```python
soup = BeautifulSoup(data, 'html.parser')
product_cards = soup.find_all('li','col-xs-2-4')

records = []
for card in product_cards:
    name_element = card.find('div','DgXDzJ rolr6k Zvjf4O').text.strip()
    product_name = str(name_element.encode('ascii','ignore'),'utf-8')
    product_price = card.find('div','bPcAVl IWBsMB').text.strip()
    sales_element = card.find('div','OwmBnn eumuJJ')
    if sales_element is None:
        product_sold = None
    else:
        product_sold = sales_element.text.strip()
    product_place = card.find('div','JVW3E2').text.strip()
    category_link = 'https://shopee.com.my/Home-Living-cat.11001155'
    product_category = '-'.join(category_link.split('/')[-1].split('-')[:-1])
    subcat_link = "https://shopee.com.my/Bedding-cat.11001155.11001165"
    product_subcategory = '-'.join(subcat_link.split('/')[-1].split('-')[:-1])

    product_info = (product_name, product_price, product_sold, product_place, product_category, product_subcategory)

    records.append(product_info)
```

```python
url = "https://shopee.com.my/Bedding-cat.11001155.11001165"
# Split the URL by '/' and take the last element
last_part = url.split('/')[-1]
# Split the last part by '-' and take the first element
result = '-'.join(last_part.split('-')[:-1])
print(result)
```

```python
# Save into CSV file
col = ['Product_Name','Product_Price','Product_Sold','Product_Place','Product_Category','Product_Subcategory']
scrape_data = pd.DataFrame(records,columns=col)

df = pd.read_csv('C:/Users/Acer/Downloads/shopee_data.csv')

# Concatenate 'scrape_data' dataframe with 'df'
df = pd.concat([df, scrape_data], ignore_index=True)

# Write the updated dataframe back to 'shopee_data.csv'
df.to_csv('C:/Users/Acer/Downloads/shopee_data.csv', index=False)
```
